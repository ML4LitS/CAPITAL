```python
import spacy
from spacy.matcher import Matcher

path_floret_model = '/home/stirunag/work/github/source_data/floret_embeddings/en_floret_model'
nlp = spacy.load(path_floret_model)
matcher = Matcher(nlp.vocab)

```

```python
# Define patterns
patterns = {
    "ARRAYEXPRESS": [{"TEXT": {"REGEX": "^E\-[A-Z]{4}\-[0-9]+$"}}],
    "ALPHAFOLD": [
        {"TEXT": {"REGEX": "^AF\-[OPQ][0-9][A-Z0-9]{3}[0-9]+\-[F][0-9]$"}},
        {"TEXT": {"REGEX": "^AF\-[A-NR-Z][0-9]([A-Z][A-Z0-9]{2}[0-9])+\-F[0-9]$"}}
    ],
    "BRENDA": [
        {"TEXT": {"REGEX": "^[0-9]+(\.[0-9]+)*\.\-$"}},
        {"TEXT": {"REGEX": "^BTO: ?[0-9]{7}$"}}
    ],
    "BIOIMAGE_ARCHIVE": [{"TEXT": {"REGEX": "^S\-BIAD[0-9]+$"}}],
    "BIOMODELS": [
        {"TEXT": {"REGEX": "^(BIOMD|MODEL)[0-9]{10}$"}},
        {"TEXT": {"REGEX": "^BMID[0-9]{12}$"}}
    ],
    "BIOSAMPLE": [{"TEXT": {"REGEX": "^SAM[NED][A-Z]?[0-9]+$"}}]
}

# Add patterns to the matcher
for key, pattern_list in patterns.items():
    matcher.add(key, *pattern_list)  # Unpack the list of patterns

```

```python
# Test on a sample text
text = "The experiment ID is E-ABCD-1234, the AlphaFold ID is AF-O1A2B345-F6, and the BioModels ID is BMID123456789012."
doc = nlp(text)

matches = matcher(doc)
for match_id, start, end in matches:
    span = doc[start:end]
    print(f"{nlp.vocab.strings[match_id]}: {span.text}")
```
